```python
import random
from bloom_filter import BloomFilter
import matplotlib.pyplot as plt

class BaseAlg:
    stream = []
    results = {}
    
    def algInit(self, size):
        self.results['found'] = []
        self.x = 0
        self.size = size
        self.bloom = BloomFilter(max_elements=size)
    
    def algRun(self):
        if (self.x in self.bloom):
            self.results['found'].append(True)
        else:
            self.results['found'].append(False)
            self.bloom.add(self.x)

    def validate(self):
        if (len(self.results['found'])!=len(stream)):
            print(f"results['found'] has the wrong size")
        else:
            foundL = []
            appeared = []
            for x in self.stream:
                foundL.append(x in appeared)
                appeared.append(x)
            Fails = 0
            sz = len(self.stream)
            for i in range(sz):
                if (self.results['found'][i]!=foundL[i]):
                    Fails += 1
            self.error = Fails/sz*100
            #print(f" Size = {self.size:.0f} Fails = {Fails} ({self.error:.2f}%)")
            #print(f"expected_repetition = {expected_repetition:.1f}")

class StreamAlg(BaseAlg):
    
    def __init__(self, stream, size=6):
        self.stream = stream
        self.algInit(size)
        self.exec()
        self.validate()

    def exec(self):
        for v in self.stream:
            self.x = v
            self.algRun()
        #for key,value in self.results.items():
        #    print(f"streaming_{key}: {value}")

N = 200

percL = []
for ii in range(5,31):
    perc = ii/100
    percL.append(perc)

errorsAll = []
for K in [1,2,3]:
    errorL = []
    M = 100 * K
    stream = [random.randint(0,N) for _ in range(M)] + [2*N] + [random.randint(0,N) for _ in range(M)]
    size_of_stream = 2*M + 1
    print(f"size_of_stream: {size_of_stream}")
    errorL = []
    for perc in percL:
        size = int(perc * size_of_stream)
        SA = StreamAlg(stream, size)
        errorL.append(SA.error)
    errorsAll.append(errorL)
    plt.plot(percL,errorL, label=f"size={size_of_stream}")
plt.legend()
```
