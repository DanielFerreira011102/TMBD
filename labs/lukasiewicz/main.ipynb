```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)

a = 1.0 / (1.0 + np.exp(-x))
b = 1.0 / (1.0 + np.exp(-x))

and_ = np.maximum(0, a + b - 1)

or_ = np.minimum(1, a + b)

implies = np.minimum(1, 1 - a + b) # not(a) or b

plt.plot(x, a, label='a')
plt.plot(x, b, label='b')
plt.plot(x, and_, label='and')
plt.plot(x, or_, label='or')
plt.plot(x, implies, label='implies')
plt.legend()
```
